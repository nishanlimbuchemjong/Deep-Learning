```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv('dataset/placement.csv')
dataset.head()
```

```python
# showing the data on graph
sns.scatterplot(x='cgpa', y='resume_score', data=dataset, hue='placed')
plt.show()
```

```python
# separating input features and output
X = dataset.iloc[:, :-1]
y = dataset['placed']
```

```python
# getting the train, testing data
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
# creating a model using perceptron
from sklearn.linear_model import Perceptron

pr = Perceptron()
pr.fit(X_train,y_train)
```

```python
# checking accuracy for checking over fitting or not

# for train dataset
pr.score(X_train, y_train)*100
```

```python
# for test dataset
pr.score(X_test, y_test)*100
```

```python
# training accuracy is less than test accuracy, so there is some problem, so we change the value of 'random_state'
```

```python
# after changing the value of random_state from 42 to 10, we get 77.5 and 75.0 accuracy of training and testing accuracy respectively
```

```python
# checking how the model is making a line i.e. how the model is making decision_region

plt.figure(figsize=(4, 3))   # for making smaller size of the figure
plot_decision_regions(X.to_numpy(), y.to_numpy(), clf=pr)
plt.show()
```

```python
# we can see that the decision region line is not drawn perfectly so we can use different hyperparameter in model obj:
# like 
# pr = Perceptron(alpha=1)
# but still the figure is same so we can say that the single perceptron does not work properly, we need to use multi layer perceptron
```
